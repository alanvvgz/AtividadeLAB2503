```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = {
    'Fortnite': 12.3,
    'CF': 8,
    'Roblox': 4,
    'PUBG': 3.23,
    'Mine': 1.4,
    'LostArk': 1.32,
    'CS': 1.3,
    'Dota2': 1.29,
    'ER': 1.29
}
series = pd.Series(dados)
```

```python
series.plot()
```

Podemos ver claramente quais jogos têm os maiores picos de jogadores online, sendo Fortnite o líder com uma grande diferença em relação aos outros jogos. O gráfico também pode indicar tendências de mercado, como quais tipos de jogos estão atualmente mais populares com base no número de players online.
